# Umsetzung: CUSUM

* CUSUM kann Änderungen in Zeitreihen feststellen. 
* Der Algorithmus baut dabei auf der *kummulierten* Summe der beobachteten Werte auf.

* Bei einer endlichen Zeitreihe $x$ mit Werten von $1, \dots, n$ starten wir mit $S_{+,1} = x_1 $. Nachfolgend wird für jeden Wert der Zeitreihe der nächste Wert von $S_{+,i} $ bestimmt zu:

$$S_{+,i}  = max(0, x_i+S_{+,i-1})$$


```python
import numpy as np
data = np.loadtxt('data/Rig Loads-data-2021-01-26 20 37 34.csv', skiprows=1, delimiter=',')
j2 = (data[:,1])
j2_norm = (j2 - j2.mean()) / j2.std() # Z-Normalisierung, dafür haben wir heute leider keine Zeit
```

```python
for i, v in enumerate(j2_norm):
    if i == 0:
        S_plus  = [v]
    else:
        S_plus.append(max(0,S_plus[-1]+v))
```

```python
fig, ax = plt.subplots(1, 1, figsize=half_plot_size)

plt.plot(j2, label='J2')
plt.plot(S_plus, label='S_plus')
plt.legend();
```

# Umsetzung: Multiple Lineare Regression

<img style="float: right;" src="img/tafel_reg.png" width=20%>


* Bei der multiplen linearen Regression wird folgender *Schätzer* definiert:

$$\hat{f}(X) = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_p X_p$$

* $X$ sind verschiedene unabhängige Variablen; $Y$ soll vorhergesagt werden. 
    * z.B.: $Y$ entspricht dem Volumenstrom einer Pumpe
    * und $X$ sind Messwerte zur Drehzahl und Stromaufnahme


* $\beta$ sind die Koeffizienten, die zunächst bestimmt werden müssen.
* Dafür wird das Verfahren der kleinsten Quadrate angewendet:  

$$min(\sum^n_{i=1} (y_i-\beta x_i)^2)$$

* Welches schließlich überführt werden kann in: 

$$\beta= (X^TX)^{−1}X^TY$$

* Einfaches, schnell umsetzbares Verfahren
* Nicht lineare Terme können einfach durch Transformation der Daten (z.B: $X_i \leftarrow X_i^2$) erprobt werden.

# Umsetzung: K-Means Clustering

<img style="float: right;" src="img/tafel_unsuper.png" width=20%>

$$W(C_k) = \frac{1}{\mid{C_k}\mid}\sum_{i, i^´}\sum_{j=1}^p(x_{ij}-x_{i^´j})^2$$

* $K$ entspricht der Anzahl von Clustern. $C_1, \dots, C_k$ entspricht den jeweiligen Clustern, d.h. einer Menge von Datensätzen. 
* $W(C)$ ist die Varianz innerhalb eines Cluster in unserem Fall berechnet über die Eulerdistanz. Sie soll minimiert werden.

1. Weise jedem Punkt einen zufälligen Cluster zu!
2. In der Folge wird iteriert:
    1. Berechne die Mitte (Means) jedes Clusters
    2. Weise jedem Punkt den Cluster zu, dessen Mitte er am nächsten ist


```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn import cluster, datasets, mixture
from ipywidgets import interact, fixed
import pandas as pd

def closest_centroid(points, centroids):
    """returns an array containing the index to the nearest centroid for each point"""
    distances = np.sqrt(((points - centroids[:, np.newaxis])**2).sum(axis=2))
    return np.argmin(distances, axis=0)

def plot_kmeans_over_iteration(iteration=0):
    fig, ax = plt.subplots(1, 1, figsize=half_plot_size)
    
    df_plot = df.copy()
    for i in range(iteration):
        means = df_plot.groupby('Cluster').mean()
        df_plot['Cluster'] = closest_centroid(df_plot[['X1', 'X2']].values, means[['X1', 'X2']].values)
    
    
    plt.scatter(df_plot['X1'], df_plot['X2'], c=df_plot['Cluster'].values, cmap=cmap_kiel, s=50)
    try:
        plt.scatter(means.X1, means.X2, cmap=cmap_kiel,c=[0, 1, 2], s=1000, marker='o', edgecolors='b')
    except: 
        pass


    plt.title('Blobs: K-Means Iteration: %s' %iteration)
    plt.ylabel('$x_2$')
    plt.xlabel('$x_1$')
    plt.tight_layout()
    plt.show()
    

centers = [[0, 1], [-1, -1], [1, -1]]
n_samples = 1500
X, _ = datasets.make_blobs(n_samples=n_samples, centers=centers, cluster_std=0.5, random_state=0)
X = StandardScaler().fit_transform(X)
y_guess = np.random.randint(3, size=n_samples)
df = pd.DataFrame({'X1': X[:, 0], 'X2':X[:, 1], 'Cluster':y_guess})
    
interact(plot_kmeans_over_iteration, iteration=(0, 10));
```

# Umsetzung: k-Neirest Neighbour

<center><img style="float: right;" src="img/knn.png" width=2% class="stretch"></center>

* Für einen Datensatz $x_0$ suchen wir die zugehörige Klasse.
* Wir wählen eine Zahl $N$ und suchen die $N$ Punkte in den Daten, die am nächsten an $x_0$ sind.
* Wir mitteln die Klassengehörigkeit aus den $N$ Punkten:

$$p(Y=y_0|X=x_0) = \frac{1}{N}\sum{I(y_i=j)}$$

```python
df = pd.read_csv('Auto.csv')
df['horsepower'] = np.where(df['horsepower']=='?', None, df['horsepower'])
df = df.dropna()
df['horsepower'] = pd.to_numeric(df['horsepower'])
df['weight'] = pd.to_numeric(df['weight'])

df[['horsepower', 'mpg']] = MinMaxScaler().fit_transform(df[['horsepower', 'mpg']])

from sklearn.neighbors import KNeighborsClassifier
model_knn_2 = KNeighborsClassifier(n_neighbors=3).fit(df[['mpg', 'horsepower']], df['origin'])
df['prediction(origin)'] = model_knn_2.predict(df[['mpg', 'horsepower']])

fig, ax = plt.subplots(1, 1, figsize=half_plot_size)

x_min, x_max = df['mpg'].min(), df['mpg'].max()
y_min, y_max = df['horsepower'].min(), df['horsepower'].max()
n = 200  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, (x_max-x_min)/n), np.arange(y_min, y_max, (y_max-y_min)/n))
Z = model_knn_2.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
#plt.pcolormesh(xx, yy, Z, cmap = cmap_light)
plt.contourf(xx, yy, Z, cmap=cmap_kiel, alpha = 0.5)

# Plot also the training points
for i in range(3):
    plt.scatter(df[df['origin'] == i+1].mpg, df[df['origin'] == i+1].horsepower, marker=filled_markers[i], label='Origin = ' + str(i+1))

ax.legend()
#ax.axis('equal')
plt.title('KNN = 3')
plt.ylabel('horsepower')
plt.xlabel('MPG')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.show()
```
